```python
# Economics concept: Supply and Demand (Claude)
# Let's represent a simple market in Python

price = 10          # price of a good in dollars
quantity_demanded = 100   # how many people want it at this price
quantity_supplied = 80    # how many sellers produce at this price

# Is this market in equilibrium?
if quantity_demanded > quantity_supplied:
    print("Shortage! Demand exceeds supply.")
    print("Price will rise.")
elif quantity_supplied > quantity_demanded:
    print("Surplus! Supply exceeds demand.")
    print("Price will fall.")
else:
    print("Equilibrium! Market clears.")
```

# Block 3 - Coding Economics Concepts in Python

```python
# How much did the price change?

original_price = 10
new_price = 150

percent_change = (( new_price - original_price)/ original_price) * 100

print(f"Original price: ${original_price}")
print(f"New price: ${new_price}")
print(f"Price change: {percent_change}%")

if percent_change > 0:
    print("Price increased, quantity demanded will fall.")

else:
    print("Price decreased, quantity demanded will rise.")
```

```python
# Supply and Demand Summary

good = "Gum"
price = 3.60
quantity_demanded = 350
quantity_supplied = 350

print(f"Market: {good}")
print(f"Current price: ${price}")
print(f"Quantity demanded: {quantity_demanded}")
print(f"Quanrtity supply: {quantity_supplied}")
print("---") 

diff = quantity_demanded - quantity_supplied

if quantity_demanded > quantity_supplied :
    print(f"Shortage of {diff} units.")
    print("Sellers will raise price")
    
elif quantity_supplied > quantity_demanded:
    print(f" Surplus of {diff} units.")
    print("Sellers will lower price")

else:
    print("Market is in equilibrium.")
    print("No presure on prices")
```
